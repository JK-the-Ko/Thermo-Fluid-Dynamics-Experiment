**Pandas DataFrame Example**
* Python Dictionary

```python
x = np.arange(0, 2 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
sinh = np.sinh(x)
cosh = np.cosh(x)
```

```python
value = dict()
value["sine"] = sin
value["cosine"] = cos
value["hyperbolic sine"] = sinh
value["hyerbolic cosine"] = cosh
```

```python
df = pd.DataFrame(data = value, index = x)
```

**plt (pyplot) Method**
* Plot Graph
  * plot(x, y, label = "label_name") → Plot Line Graph
  * scatter(x, y, marker = marker_type, label = "label_name") → Plot Scatter Graph
  * bar(x, y, label = "label_name") → Plot Bar Chart
* Indicator
  * xlabel("x-axis name")
  * ylabel("y-axis name")
  * legend(loc = "location")
* Show Graph
  * show() → Show Graph
* Save Graph
  * savefig(fname = "location", dpi = dpi_num)

**Plot Graph from Numpy Array**

```python
x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y, label = "Line Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, marker = ".", label = "Scatter Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, marker = "*", label = "Scatter Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, marker = "s", label = "Scatter Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
plt.bar(x, y, label = "Bar Chart")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
