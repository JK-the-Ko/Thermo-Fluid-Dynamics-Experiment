* Bubble Sorting
  * Time Complexity → O(N^2)

https://www.youtube.com/watch?v=xli_FI7CuzA

```python
def bubble_sort(input_list) :
  list_len = len(input_list)
  return_list = list(input_list)
  for i in range(list_len - 1) :
    for j in range(i + 1, list_len) :
      if return_list[j] < return_list[i] :
        current_value = return_list[i]
        return_list[i] = return_list[j]
        return_list[j] = current_value
  
  return return_list  
```

```python
list_1 = [10, 0, 1, 9, 8, -1, 5]
list_new = bubble_sort(list_1)
list_new
```

* Attribute and Method
  * Attribute → Variable
  * Method → Function

```python
class cartesian_coordinate :
  def __init__(self, x, y, z) : 
    self.x = x
    self.y = y
    self.z = z
        
  def move(self, delta) :
    self.x += delta
    self.y += delta
    self.z += delta

  def get_coordinates(self) :
    return self.x, self.y, self.z
```

```python
class calculator :
  def __init__(self, num_1, num_2) :
    self.num_1 = num_1
    self.num_2 = num_2
  
  def add(self) :
    return self.num_1 + self.num_2
  
  def subtract(self) :
    return self.num_1 - self.num_2
  
  def multiply(self) :
    return self.num_1 * self.num_2
  
  def divide(self) :
    return self.num_1 / self.num_2
```

* Example 2 - Name List Tool

```python
class student_name_list :
  def __init__(self) :
    self.name_list = list()

  def add_name(self, name) :
    self.name_list.append(name)

  def get_name(self) :
    for name in self.name_list :
      print(name)

  def remove_name(self, name) :
    if name in self.name_list :
      self.name_list.remove(name)
      print(f"{name} removed!")
    
    else :
      print(f"{name} not in list!")
  
  def get_student_number(self) :
    print(len(self.name_list))
```

**Object-Oriented Programming (OOP)**

객체 지향 프로그래밍(영어: Object-Oriented Programming, OOP)은 컴퓨터 프로그래밍의 패러다임 중 하나이다. 객체 지향 프로그래밍은 컴퓨터 프로그램을 명령어의 목록으로 보는 시각에서 벗어나 여러 개의 독립된 단위, 즉 "객체"들의 모임으로 파악하고자 하는 것이다. 각각의 객체는 메시지를 주고받고, 데이터를 처리할 수 있다.

* Polymorphism

```python
class integer_operation :
  def __init__(self, num_1, num_2) :
    self.num_1 = num_1
    self.num_2 = num_2
  
  def add(self) :
    return self.num_1 + self.num_2
```

```python
class list_operation :
  def __init__(self, list_1, list_2) :
    self.list_1 = list_1
    self.list_2 = list_2

  def add(self) :
    list_new = list()
    for i in range(len(self.list_1)) :
      list_new.append(self.list_1[i] + self.list_2[i])
    return list_new
```

```python
io = integer_operation(10, 11)
io.add()
```

```python
lo = list_operation([1,2,3], [4,5,6])
lo.add()
```

* Inheritance

```python
class Polygon :
  def __init__(self, num_edge) :
    self.num_edge = num_edge
    
  def count_edge(self) :
    print(f"This Polygon Has {self.num_edge} Edges")
```

```python
class Triangle(Polygon) :
  def __init__(self, num_edge) :
    super(Triangle, self).__init__(num_edge)
```

```python
triangle = Triangle(3)
triangle.count_edge()
```
