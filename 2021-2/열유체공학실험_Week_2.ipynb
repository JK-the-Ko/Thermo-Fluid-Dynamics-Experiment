**if**

if → elif → else

**Basic Structure**

if condition_1 :

elif condition_2 :

elif condition_3 :

else :

```python
grade = "A"

if grade == "F" :
  print("F")

elif grade == "D" :
  print("D")

elif grade == "C" :
  print("C")

elif grade == "B" :
  print("B")

elif grade == "A" :
  print("A")
```

```python
money, price = 100, 200

if money > price :
  print("Please keep the change")

elif money == price :
  print("Thank you")

else :
  print("Give me more money")
```

* Nested for loop

```python
list_2d = [[1, 2, 3], [4, 5, 6]]
```

```python
for i in range(2) :
  for j in range(3) :
    print(list_2d[i][j])
```

```python
list_2d = list()

for i in range(2) :
  list_1d = list()
  
  for j in range(3) :
    list_1d.append(i + j)
  
  list_2d.append(list_1d)
```

```python
list_2d
```

**while loop**

**Basic Structure**

while "condition" :

...

```python
i = 0
while i < 10 :
  print(i)
  i += 1
```

```python
boolean = True

while boolean :
  char = input()
  
  if char == "stop" :
    boolean = False
 
  else :
    print(char)
```
